```python
from typing_extensions import dataclass_transform
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("energy.csv")
print(data)
#Andhra
#karnataka
#Line chart
srh=data[data['State']=='Andhra']
rcb= data[data['State']=='Karnataka']
#line chart
fig,diag = plt.subplots(2)
diag[0].plot(srh['Date'],srh['Consumption'],color='yellow')
diag[1].plot(rcb["Date"],rcb['Consumption'],color='orange')

srh=data[data['State']=='Telengana']
rcb= data[data['State']=='Tripura']
#line chart
fig,diag = plt.subplots(2)
diag[0].bar(srh['Date'],srh['Consumption'],color='red')
diag[1].bar(rcb["Date"],rcb['Consumption'])
#pie chat
state1='Kerala'
state2='Delhi'
region=[state1,state2]
srh=data[data['State']=='Kerala']['Consumption'].sum()
rcb= data[data['State']=='Delhi']['Consumption'].sum()
regiondata=[srh,rcb]
fig,diag = plt.subplots(2)
diag[0].pie(regiondata,labels=region,autopct='%1.1f%%')
```
